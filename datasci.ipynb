```
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('heart_data.csv')
 
xaxis = ['50< smoke','60< smoke', '70< smoke']
yaxis = [0,0,0]
 
for label,row in data.iterrows():
  if row["age"] <=50 and row["smoking"]==1:
    yaxis[0]+=1
  elif row["age"] in range(50,60) and row["smoking"] ==1:
    yaxis[1]+=1
  elif row["age"] in range(60,70) and row["smoking"] ==1:
    yaxis[2]+=1
 
plt.boxplot(xaxis,yaxis)
```

```
import pandas as pd
smokers=0
death_smoke=0
heart_data=pd.read_csv('heart_data.csv')
for label,row in heart_data.iterrows():
    if(row['smoking']==1):
        smokers= smokers+1
    if(row['DEATH_EVENT']==1 and row['smoking']==1):
        death_smoke=death_smoke+1       
print("probability of smoker will die of heart attack",death_smoke/smokers)
```

```
import numpy as np
np.random.seed(6)
sample_chance=np.random.choice(a= heart_data['DEATH_EVENT'], size=200)
print ("Sample mean:", sample_chance.mean() )
print("Population mean:", heart_data['DEATH_EVENT'].mean())
import scipy.stats as stats
import math
np.random.seed(10)
sample_size = 200
sample = np.random.choice(a= heart_data['DEATH_EVENT'],size = sample_size)
sample_mean = sample.mean()
z_critical = stats.norm.ppf(q = 0.95)
print("z-critical value: ",z_critical)
pop_stdev = heart_data['DEATH_EVENT'].std()
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error,sample_mean +
margin_of_error)
print("Confidence interval:",end=" ")
print(confidence_interval)
print("True mean: {}".format(heart_data['DEATH_EVENT'].mean()))
```

Importing data


```
import numpy as np
import pandas as pd

df = pd.read_csv('heart_data.csv')
```

```
df.head()
```

spliting data


```
X=df[['sex', 'smoking','anaemia', 'diabetes', 'high_blood_pressure', 'age', 'ejection_fraction', 'time',
     'platelets', 'creatinine_phosphokinase', 
      'serum_creatinine', 'serum_sodium']].values
y=df[['DEATH_EVENT']].values
```

Random forest algorithm
1. train and test sets


```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, np.array(y).ravel(), random_state=1,test_size=0.2)
```

Building model

```
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score

List_estim=[10,100,200]
List_depth=[2,10,15]
List_leaf_nodes=[10,20,40]
for a in List_estim:
    for b in List_depth:
        for c in List_leaf_nodes:
            clf = RandomForestClassifier(n_estimators = a, max_depth = b, max_leaf_nodes = c).fit(X_train, y_train)

            pred = clf.predict(X_test)
            
            print("Model parameters n_estimators:%0.0f, max_depth:%0.0f, max_leaf_nodes:%0.0f," % (a,b,c))
            print('Accuracy Score : {:.2f}%'.format(clf.score(X_test, y_test)*100))
            print('\nf1 Score : ', f1_score(y_test, pred))
            print('\nPrecision Score : ', precision_score(y_test, pred))
            print('\nRecall Score : ', recall_score(y_test, pred))
            print('-------------------------------------' )
```
